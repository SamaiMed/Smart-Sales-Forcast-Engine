```python
import numpy as np
import tensorflow as tf
from string import punctuation
from collections import Counter
import collections
import random
import math
from six.moves import xrange 
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output
import csv
from tensorflow.examples.tutorials.mnist import input_data
from tensorflow.contrib.layers import fully_connected
```

```python
seq_len=30
embeding_size=30
n_steps = seq_len
n_inputs = embeding_size
n_neurons = 1
n_outputs = 1
learning_rate = 0.0001
nbr_layers=1
```

```python
with open('lab.txt', 'r') as f:
    labels_org = f.read()
df=pd.read_csv('Seq_Embo_f.csv',header=None)
features=np.array(df[:][:]).astype(np.float32)
labels = np.array([int(_)  for _ in labels_org.split()]).astype(np.float32)
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y=train_test_split(features,labels,test_size=0.2)

train_x=train_x.reshape(-1,30,30)
test_x=test_x.reshape(-1,30,30)


print("\t\t\tFeature Shapes:")
print("Train set: \t\t{}".format(train_x.shape), "\nTest set: \t\t{}".format(test_x.shape))
print("label set: \t\t{}".format(train_y.shape), "\nTest label set: \t\t{}".format(test_y.shape))
```

```python
def get_batches(x, y, batch_size=800):
    n_batches = len(x)//batch_size
    x, y = x[:n_batches*batch_size], y[:n_batches*batch_size]
    for ii in range(0, len(x), batch_size):
        yield x[ii:ii+batch_size], y[ii:ii+batch_size]
```

```python
tf.reset_default_graph()
X = tf.placeholder(tf.float32, [None, n_steps, n_inputs])
y = tf.placeholder(tf.float32, [None])
keep_drop=tf.placeholder(tf.float32, name="keep_prob")
def Cell():
    basic_cell = tf.contrib.rnn.GRUCell(num_units=n_neurons)
    drop_cell=tf.contrib.rnn.DropoutWrapper(basic_cell, output_keep_prob=keep_drop)
    return basic_cell
temp=[Cell() for _ in range(nbr_layers)]
cells=tf.contrib.rnn.MultiRNNCell(temp)
outputs, states = tf.nn.dynamic_rnn(cells, X, dtype=tf.float32)
"""
for one_lstm_cell in temp:
    one_kernel, one_bias = one_lstm_cell.variables
    # I think TensorBoard handles summaries with the same name fine.
    tf.summary.histogram("Kernel", one_kernel)
    tf.summary.histogram("Bias", one_bias)
"""
pp = fully_connected(outputs[:,-1],n_outputs, activation_fn=None)
logits=tf.nn.tanh(pp)
#xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logits)
#loss = tf.reduce_mean(xentropy)
loss=tf.reduce_mean(y-logits)


optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
training_op = optimizer.minimize(loss)
#correct = tf.nn.in_top_k(logits, y, 1)
#accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
total_error = tf.reduce_sum(tf.square(y - tf.reduce_mean(y)))
unexplained_error = tf.reduce_sum(tf.square(y - logits))
accuracy = 1- tf.div(total_error, unexplained_error)


tf.summary.scalar('accuracy', accuracy)
tf.summary.scalar('loss', loss)
merged = tf.summary.merge_all()
init = tf.global_variables_initializer()
```

```python
n_epochs = 10
saver = tf.train.Saver()
with tf.Session() as sess:
    init.run()
    print('Initialized')
    train_writer = tf.summary.FileWriter('train_III', sess.graph)
    for epoch in range(n_epochs):
        act_1=0
        act_2=0
        for ii, (xx, yy) in enumerate(get_batches(train_x, train_y, 800), 1):
            _,acc_train,pred,ll=sess.run([training_op,accuracy,logits,loss], feed_dict={X: xx, y: yy, 
                                                                                        keep_drop: 1})
            act_1+=acc_train
            act_2+=acc_train
            if ii%1==0:
                print('Made-->',ii,act_2/1)
                summary=merged.eval(feed_dict={X: xx, y: yy, keep_drop: 1})
                train_writer.add_summary(summary, ii)
                act_2=0
        acc_test = accuracy.eval(feed_dict={X: test_x, y: test_y, keep_drop: 1})
        print(epoch, "Train accuracy:", act_1/(ii), "Test accuracy:", acc_test,'loss->',ll)
        plt.plot(pred[:100],color='red')
        plt.plot(train_y[:100],color='b')
        plt.show()
        saver.save(sess, "checkpoints/sento-anal_fun.ckpt")

"""
    from matplotlib import pyplot as plt
    o=outputs.eval(feed_dict={X: X_batch, y: y_batch})
    l=states.eval(feed_dict={X: X_batch, y: y_batch})
    for i in range(10):
        plt.imshow(o[i,:,:])
        plt.show()
"""
```
